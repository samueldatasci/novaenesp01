```python
import qrcode
from PIL import Image
import io
```

```python

def generate_qr_with_logo(link, logo_buffer):
    # 1. Setup QR code parameters
    # High error correction (Level H) is REQUIRED to allow for a center logo
    qr = qrcode.QRCode(
        version=1,
        error_correction=qrcode.constants.ERROR_CORRECT_H,
        box_size=10,
        border=4,
    )
    qr.add_data(link)
    qr.make(fit=True)

    # Create the QR image (black and white)
    qr_img = qr.make_image(fill_color="black", back_color="white").convert('RGB')

    # 2. Prepare the Logo (the Matplotlib figure buffer)
    logo = Image.open(logo_buffer)
    
    # Calculate dimensions (logo should be ~20% of QR size)
    qr_width, qr_height = qr_img.size
    logo_size = qr_width // 4  # Adjusting for visibility
    logo = logo.resize((logo_size, logo_size))

    # 3. Paste logo into the center
    pos = ((qr_width - logo_size) // 2, (qr_height - logo_size) // 2)
    qr_img.paste(logo, pos)

    return qr_img

```

```python
# Usage in your Streamlit app:
buf = None
qr_code_final = generate_qr_with_logo("https://novaenesp01.streamlit.app/", buf)
qr_code_final.save("qr_with_logo.png")
#st.image(qr_code_final, caption="Scan to visit the app!")
```
